```python
import os
import glob

prediction_folder = '/home/hyungyu/Experiments/yolov5/FINAL2_Val_V2_Exp_S_Noagono/Batch_2_Yolov5s/labels'
# prediction_folder = '/home/hyungyu/Experiments/yolov5/FINAL_Val_V2_Experiments_S/Batch_2_Yolov5s_Conf_0.4_noagnostic/labels'
gt_folder = '/home/hyungyu/Experiments/yolov5/yolov5_v2_dataset2/val/labels'

def count_hairs(txt_file):
    count = 0
    with open(txt_file, 'r') as file:
        for line in file:
            class_id = int(line.split()[0])
            count += class_id + 1  # 클래스에 1을 더해야 모발의 유형과 같아짐
    return count

# 오차율 계산 함수
def calculate_error_rate(predicted, actual):
    if actual == 0:
        return 0  # actual hair counting이 0이면 오차율 0.
    return abs(predicted - actual) / actual * 100

def calculate_error(predicted, actual):
    if actual == 0:
        return 0
    return abs(predicted - actual)


pred_counts = {} # model이 predict한 모발 개수 counting
for txt_file in glob.glob(os.path.join(prediction_folder, '*.txt')): 
    file_name = os.path.basename(txt_file)
    pred_counts[file_name] = count_hairs(txt_file)


gt_counts = {} # ground truth에서 모발 개수 counting
for txt_file in glob.glob(os.path.join(gt_folder, '*.txt')):
    file_name = os.path.basename(txt_file)
    gt_counts[file_name] = count_hairs(txt_file)
    
error_rates = {}
error_counts ={}
for file_name in sorted(gt_counts.keys(), key=lambda x: x.split('_')[0]):
    if file_name in pred_counts:
        error_rate = calculate_error_rate(pred_counts[file_name], gt_counts[file_name])
        error_rates[file_name] = error_rate
        
        error_count = calculate_error(pred_counts[file_name], gt_counts[file_name])
        error_counts[file_name] = error_count
        print(f"{file_name} - 예측된 모발 개수: {pred_counts[file_name]}, 실제 모발 개수: {gt_counts[file_name]}, 오차 개수 : {error_count} 개, 오차율: {error_rate:.2f}%")
        # print(f"{file_name} - 예측된 모발 개수: {pred_counts[file_name]}, 실제 모발 개수: {gt_counts[file_name]}, 오차율: {error_rate:.2f}%")
    else:
        print(f"{file_name} - 예측 결과 없음.")

```

```python
total_error_rate = sum(error_rates.values())
average_error_rate = total_error_rate / len(error_rates) if error_rates else 0

total_error_count = sum(error_counts.values())
average_error_count = total_error_count/ len(error_counts)

# print(f"총 오차율: {total_error_rate:.2f}%")
print(f"평균 오차율: {average_error_rate:.2f}%")
print(f"평균 오차 개수: {average_error_count:.2f} 개")
```

```python
import os
import glob

prediction_folder = '/home/hyungyu/Experiments/yolov5/FINAL2_Val_V2_Exp_S_Noagono/Batch_32_Yolov5s/labels'
# prediction_folder = '/home/hyungyu/Experiments/yolov5/FINAL_Val_V2_Experiments_S/Batch_2_Yolov5s_Conf_0.4_noagnostic/labels'
gt_folder = '/home/hyungyu/Experiments/yolov5/yolov5_v2_dataset2/val/labels'

def count_hairs(txt_file):
    count = 0
    with open(txt_file, 'r') as file:
        for line in file:
            class_id = int(line.split()[0])
            count += class_id + 1  # 클래스에 1을 더해야 모발의 유형과 같아짐
    return count

# 오차율 계산 함수
def calculate_error_rate(predicted, actual):
    if actual == 0:
        return 0  # actual hair counting이 0이면 오차율 0.
    return abs(predicted - actual) / actual * 100

def calculate_error(predicted, actual):
    if actual == 0:
        return 0
    return abs(predicted - actual)

def test_density(predicted, actual):
    return predicted/actual


pred_counts = {} # model이 predict한 모발 개수 counting
for txt_file in glob.glob(os.path.join(prediction_folder, '*.txt')):
    file_name = os.path.basename(txt_file)
    pred_counts[file_name] = count_hairs(txt_file)


gt_counts = {} # ground truth에서 모발 개수 counting
for txt_file in glob.glob(os.path.join(gt_folder, '*.txt')):
    file_name = os.path.basename(txt_file)
    gt_counts[file_name] = count_hairs(txt_file)
    
error_rates = {}
error_counts ={}
density = {}
for file_name in sorted(gt_counts.keys(), key=lambda x: x.split('_')[0]):
    if file_name in pred_counts:
        error_rate = calculate_error_rate(pred_counts[file_name], gt_counts[file_name])
        error_rates[file_name] = error_rate
        
        error_count = calculate_error(pred_counts[file_name], gt_counts[file_name])
        error_counts[file_name] = error_count
        
        tt_density = test_density(pred_counts[file_name], gt_counts[file_name])
        density[file_name] = tt_density
        print(f"{file_name} - 예측된 모발 개수: {pred_counts[file_name]}, 실제 모발 개수: {gt_counts[file_name]}, 오차 개수 : {error_count} 개, 밀도 : {tt_density}, 오차율: {error_rate:.2f}%")
    else:
        print(f"{file_name} - 예측 결과 없음.")

```
